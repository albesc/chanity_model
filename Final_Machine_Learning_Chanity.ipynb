### Separate the positive questions and the negative questions

```python
positively_keyed = ['EXT1', 'EXT3', 'EXT5', 'EXT7', 'EXT9',
                    'EST1', 'EST3', 'EST5', 'EST6', 'EST7', 
                    'EST8', 'EST9', 'EST10',
                    'AGR2', 'AGR4', 'AGR6', 'AGR8', 'AGR9', 'AGR10',
                    'CSN1', 'CSN3', 'CSN5', 'CSN7', 'CSN9', 'CSN10', 
                    'OPN1', 'OPN3', 'OPN5', 'OPN7', 'OPN8', 'OPN9', 
                    'OPN10']

negatively_keyed = ['EXT2', 'EXT4', 'EXT6', 'EXT8', 'EXT10',
                    'EST2', 'EST4',
                    'AGR1', 'AGR3', 'AGR5', 'AGR7', 
                    'CSN2', 'CSN4', 'CSN6', 'CSN8', 
                    'OPN2', 'OPN4', 'OPN6']
```

### Change the value of negative questions to 6 minus the original value

```python
data.loc[:, negatively_keyed] = 6 - data.loc[:, negatively_keyed]  
data
```

## **Clustering**

### Using KMeans method from scikit-learn library to cluster the data into 5 cluster

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=1)
k_fit = kmeans.fit(data)
```

###Print the cluster's labels

```python
k_fit.labels_
```

### Add the cluster's labels to *cluster* column

```python
predictions = k_fit.labels_
data['Clusters'] = predictions
data
```

### Summing up the different questions groups and find the mean values

```python
# Summing up the different questions groups
listdata = list(data)
ext = listdata[0:10]
est = listdata[10:20]
agr = listdata[20:30]
csn = listdata[30:40]
opn = listdata[40:50]

# Find the mean values for each groups
sumdata = pd.DataFrame()
sumdata['extroversion']  = data[ext].sum(axis=1) / 10
sumdata['neuroticism']   = data[est].sum(axis=1) / 10
sumdata['agreeable']     = data[agr].sum(axis=1) / 10
sumdata['conscientious'] = data[csn].sum(axis=1) / 10
sumdata['openness']      = data[opn].sum(axis=1) / 10
sumdata['clusters']      = predictions 
sumdata.groupby('clusters').mean().round(2) * 20       # times 20 to scale the value to 20-100
```

###Print the row counts for each cluster

```python
data.Clusters.value_counts()
```

### Pending section (label binarizer)

```python
# import module
from sklearn.preprocessing import LabelBinarizer

y = LabelBinarizer().fit_transform(data.Clusters)
y = pd.DataFrame(y, columns = ['0','1','2','3','4'])
data
```

## **Build a Model**

### Splits the data into training and testing

```python
from sklearn.model_selection import train_test_split
# define label
x = data.drop('Clusters', axis=1)
y.astype(int)

# split dataframe 80% train, 20% test
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=1)
print("Training size: {}".format(train_x.shape[0]))
print("Testing size: {}".format(test_x.shape[0]))
```

### Rescale the data using MinMaxScaler

```python
# rescale
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# train
train_x = scaler.fit_transform(train_x.astype(np.float64))
# test
test_x = scaler.fit_transform(test_x.astype(np.float64))
```

### Create a model

```python
# neural network
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=512, activation='relu', input_shape=(50,)))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(units=5, activation='softmax'))


model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(train_x, train_y, epochs=5,validation_data=(test_x, test_y))
```

### Evaluate the model with test data

```python
test_loss, test_accuracy = model.evaluate(test_x, test_y)
print(test_loss, test_accuracy)
```

### Predict the output from the given input to the model

```python
pred = model.predict([[2,	2,	3,	2,	2,	4,	1,	2,	2,	2,	4,	3,	4,	4,	4,	4,	2,	2,	4,	4,	4,	3,	4,	4,	3,	4,	4,	4,	4,	3,	4,	4,	4,	4,	2,	4,	4,	4,	4,	4,	5,	4,	4,	4,	3,	4,	4,	5,	5,	3]])
print(pred)
```

## **Exporting model to TFLite**

```python
export_dir = 'saved_model/1'

tf.saved_model.save(model,export_dir)
```

```python
# Select mode of optimization
mode = "Speed" 

if mode == 'Storage':
    optimization = tf.lite.Optimize.OPTIMIZE_FOR_SIZE
elif mode == 'Speed':
    optimization = tf.lite.Optimize.OPTIMIZE_FOR_LATENCY
else:
    optimization = tf.lite.Optimize.DEFAULT
```

```python
# Use the TFLiteConverter SavedModel API to initialize the converter
converter = tf.lite.TFLiteConverter.from_saved_model(export_dir)

# Set the optimzations
converter.optimizations = [optimization]

# Invoke the converter to finally generate the TFLite model
tflite_model = converter.convert();
```

```python
tflite_model_file = pathlib.Path('./model.tflite')
tflite_model_file.write_bytes(tflite_model)
```
